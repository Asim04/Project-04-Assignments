# Problem: High Low
We want you to gain more experience working with control flow and Booleans in Python. To do this, we are going to have you develop a game! The game is called High-Low and the way it's played goes as follows:

Two numbers are generated from 1 to 100 (inclusive on both ends): one for you and one for a computer, who will be your opponent. You can see your number, but not the computer's!

You make a guess, saying your number is either higher than or lower than the computer's number

If your guess matches the truth (ex. you guess your number is higher, and then your number is actually higher than the computer's), you get a point!

These steps make up one round of the game. The game is over after all rounds have been played.


```python
import random
print("High and Low Game")

round = 5

def main():

  print("Wlecome to the High Low Game")
  print("*****************************")

  your_score = 0

  for i in range(round):
    print("Round ", i + 1)

    computer_number : int = random.randint(1,100)
    your_number : int = random.randint(1,100)
    print("you Number is", your_number)

    choice: str = input("Do you think your number is higher or lower then the computer's number?")
    higher_and_correct: bool = choice == "higher" and your_number > computer_number
    lower_and_correct: bool = choice == "lower" and your_number < computer_number

    if higher_and_correct or lower_and_correct:
      print("‚úÖ  Your were right! The computer\ s number was", computer_number)
      your_score += 1

    else:
      print("‚ùå  That\'s incorrect. The computer\'s number was ", computer_number)

    print("your score is now ", your_score)
    print()

  print(" Thanks for playing! üéóÔ∏è")

if __name__ == "__main__":
  main()


```

# Problem: Planetary Weight Calculator
Milestone #1: Mars Weight
A few years ago, NASA made history with the first controlled flight on another planet. Its latest Mars Rover, Perseverance, has onboard a 50cm high helicopter called Ingenuity. Ingenuity made its third flight, during which it flew faster and further than it had on any of its test flights on Earth. Interestingly, Ingenuity uses Python for some of its flight modeling software!

Ingenuity on the surface of Mars (source: NASA)

When programming Ingenuity, one of the things that NASA engineers need to account for is the fact that due to the weaker gravity on Mars, an Earthling's weight on Mars is 37.8% of their weight on Earth. Write a Python program that prompts an Earthling to enter their weight on Earth and prints their calculated weight on Mars.

The output should be rounded to two decimal places when necessary. Python has a round function which can help you with this. You pass in the value to be rounded and the number of decimal places to use. In the example below, the number 3.1415926 is rounded to 2 decimal places which is 3.14.

```python
print("Planetary Weight Calculator")

def main():
  print("Wlecome to the planet weigth calculater")
  print("***************************************üëÄüöÄ")

  # User se Earth par weight input lena
  earth_weight = float(input("Enter your weigth on Earth (in Kg): "))

  # Mars par weight calculate karna (37.8% of Earth weight)
  mars_weight = earth_weight * 0.378

  # Result ko 2 decimal places tak round karna
  mars_weight_rounded = round(mars_weight, 2)

  # # Output print karna
  print(f"Your weight on Mars would be {mars_weight_rounded}")

if __name__ == "__main__":
  main()
```

```python
def access_element(lst, index):
    if 0 <= index < len(lst):
        return f"Element at index {index}: {lst[index]}", 10  # 10 points for success
    else:
        return f"Error: Index out of range! Valid range is 0 to {len(lst) - 1}.", 0

def modify_element(lst, index, new_value):
    if 0 <= index < len(lst):
        lst[index] = new_value
        print()
        return f"List updated! New list: {lst}", 15  # 15 points for modify
    else:
        return f"Error: Index out of range! Valid range is 0 to {len(lst) - 1}.", 0

def slice_list(lst, start, end):
    if start < 0 or end > len(lst) or start > end:
        return f"Error: Invalid indices! Start must be >= 0, end <= {len(lst)}, and start <= end.", 0
    else:
        return f"Sliced list: {lst[start:end]}", 20  # 20 points for slice

def get_initial_list():
    print("Let's create your list!")
    size = 0
    while size < 5:
        try:
            size = int(input("How many elements do you want (minimum 5): "))
            if size < 5:
                print("Please enter a number >= 5!")
        except ValueError:
            print("Invalid input! Please enter a number.")

    user_list = []
    for i in range(size):
        value = input(f"Enter element {i + 1}: ")
        user_list.append(value)

    return user_list

def main():
    # Get initial list from user
    game_list = get_initial_list()
    print(f"\nYour starting list: {game_list}")

    score = 0  # Initialize score

    while True:
        # Display menu with score
        print(f"\nScore: {score}")
        print("What would you like to do?")
        print("1. Access an element (10 points)")
        print("2. Modify an element (15 points)")
        print("3. Slice the list (20 points)")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            try:
                index = int(input(f"Enter the index to access (0-{len(game_list)-1}): "))
                result, points = access_element(game_list, index)
                print(result)
                score += points
            except ValueError:
                print("Error: Please enter a valid integer for the index!")

        elif choice == "2":
            try:
                index = int(input(f"Enter the index to modify (0-{len(game_list)-1}): "))
                new_value = input("Enter the new value: ")
                result, points = modify_element(game_list, index, new_value)
                print(result)
                score += points
            except ValueError:
                print("Error: Please enter a valid integer for the index!")

        elif choice == "3":
            try:
                start = int(input("Enter the start index: "))
                end = int(input("Enter the end index: "))
                result, points = slice_list(game_list, start, end)
                print(result)
                score += points
            except ValueError:
                print("Error: Please enter valid integers for start and end indices!")

        elif choice == "4":
            print("Thanks for playing! Final list:", game_list)
            print("Final Score:", score)
            break

        else:
            print("Invalid choice! Please select 1, 2, 3, or 4.")

if __name__ == "__main__":
    main()
```
